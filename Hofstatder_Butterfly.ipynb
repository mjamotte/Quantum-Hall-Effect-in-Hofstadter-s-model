# Hofstatder Butterfly

```python
Np = 90
Nq = 90

kx = 1
ky = 1
J = 1

all_spectre = np.array([])
all_alpha = np.array([])

for q in range(1,Nq+1):
    for p in range(1,q):
        alpha = np.ones(q)*p/q
        spectre_alpha = linalg.eigh(Fun_Hof.H_Hofstatder_tore(kx,ky,p,q))[0]
        
        all_spectre = np.concatenate((all_spectre, spectre_alpha), axis=None)
        all_alpha = np.concatenate((all_alpha, alpha), axis=None)
```

```python
fig_F = pyplot.figure(figsize=(8,8))
ax1 = pyplot.subplot(111)

ax1.plot(all_alpha,all_spectre/J,'.k',Markersize = 0.1)
    
ax1.set_xlabel('$p/q$',Fontsize=14)
ax1.set_ylabel('$E/J$',Fontsize=14)
#ax1.grid(axis='both')
```
